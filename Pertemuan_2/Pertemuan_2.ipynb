# Karakterisitik: Model Instrumen Sederhana

## 1. Konsep Dasar
Model umum dari suatu instrumen sederhana melibatkan proses fisik yang diukur serta variabel fisik teramati yang disebut sebagai **measurand**. Variabel ini tidak selalu identik dengan besaran yang ingin diukur tetapi memiliki hubungan tertentu dengannya.  
Misalnya, massa suatu objek sering diukur melalui proses penimbangan, di mana variabel yang diukur adalah gaya gravitasi yang bekerja pada massa tersebut.

## 2. Elemen Utama Instrumen
Elemen utama dalam model instrumen adalah **sensor**, yang berfungsi mengubah variabel fisik menjadi variabel sinyal. Variabel sinyal ini dapat dimanipulasi dalam sistem transmisi seperti **sirkuit listrik** atau **mekanik**, memungkinkan pengiriman data ke perangkat tampilan atau perekam yang dapat berlokasi jauh dari sensor.  
- Dalam sistem **listrik**, **tegangan** merupakan variabel sinyal yang umum digunakan.  
- Dalam sistem **mekanik**, **perpindahan** atau **gaya** sering dijadikan sebagai variabel sinyal.  

Jika keluaran sinyal dari sensor **lemah**, diperlukan **penguat** untuk meningkatkan amplitudo sinyal sebelum dikirimkan ke perangkat tampilan atau perekam.  
Instrumen modern sering menyediakan **keluaran sinyal digital**, memungkinkan integrasi dengan sistem akuisisi data berbasis komputer melalui **konverter analog-ke-digital (ADC)**.

## 3. Sensor Pasif dan Aktif
Sensor dapat dikategorikan menjadi dua jenis:  

### a. **Sensor Pasif**
- Tidak menambahkan energi ke lingkungan selama proses pengukuran.  
- Contoh: **Termokopel**, yang mengubah suhu menjadi sinyal tegangan berdasarkan gradien temperatur.  

### b. **Sensor Aktif**
- Menambahkan energi ke lingkungan pengukuran.  
- Contoh: **Radar atau sonar**, yang mengukur jarak dengan mengirimkan gelombang elektromagnetik atau akustik dan menganalisis pantulannya.

---

## 4. Kalibrasi
Kalibrasi sensor mengacu pada hubungan antara **variabel fisik yang diukur** dengan **sinyal keluarannya**.  
Langkah-langkah kalibrasi:  
1. Memberikan **input fisik** yang diketahui.  
2. Mencatat **keluaran sensor**.  
3. Menentukan hubungan antara input dan output, baik **linear** maupun **non-linear**.  

**Karakteristik kalibrasi sensor:**  
- **Sensitivitas sensor**: Ditentukan oleh kemiringan kurva kalibrasi.  
- **Batas jenuh**: Titik di mana respons sensor tidak lagi berubah meskipun input fisik meningkat.  
- **Rentang dinamis**: Perbedaan antara nilai input terkecil dan terbesar yang dapat diukur dengan andal.

## 5. Input Pengganggu dan Pemodifikasi
Dalam pengukuran, keluaran sensor dapat dipengaruhi oleh variabel selain **measurand** yang dimaksudkan.  

### a. **Input Pengganggu**
- Menyebabkan sensor memberikan respons seolah-olah dipengaruhi oleh **superposisi measurand dan input pengganggu**.  
- Contoh: **Getaran struktural** yang dapat mengganggu pengukuran gaya.  

### b. **Input Pemodifikasi**
- Mengubah karakteristik sensor itu sendiri, mempengaruhi hubungan antara **input dan output sensor**.  
- Contoh: **Perubahan suhu** yang menyebabkan perubahan kalibrasi sensor.

## 6. Akurasi dan Kesalahan Pengukuran
### a. **Akurasi Instrumen**
Akurasi didefinisikan sebagai **perbedaan antara nilai sejati measurand dan nilai yang diindikasikan oleh instrumen**.  
<img src="akurasi.png" width="250"/>

### b. **Kesalahan Pengukuran**
Kesalahan dalam pengukuran dapat dikategorikan sebagai:  
1. **Kesalahan Sistematis (Bias)**  
   - Disebabkan oleh **kalibrasi yang salah**, perubahan lingkungan, penuaan sensor, atau **kesalahan manusia**.  
2. **Kesalahan Acak (Noise)**  
   - Variasi yang **tidak dapat diprediksi**, umumnya mengikuti **distribusi Gaussian**.

## 7. Sumber Kesalahan Acak (Noise)
Kesalahan acak dapat berasal dari:  
- **Variabilitas measurand**.  
- **Gangguan lingkungan**.  
- **Noise internal sensor**.  
- **Noise dalam jalur transmisi sinyal**.  

Dalam sistem listrik, noise dapat berasal dari **interferensi jaringan listrik**, sementara dalam sistem mekanik, **gesekan** dapat mengubah propagasi sinyal.  
Metode untuk mengukur kualitas pengukuran:  
- **Rasio Sinyal terhadap Noise (SNR)**, di mana nilai **SNR tinggi menunjukkan kualitas sinyal yang baik**.

## 8. Fusi Sensor
Fusi sensor adalah **proses menggabungkan data dari beberapa sensor** untuk meningkatkan kualitas pengukuran.  
Contoh penerapan:  
- **Kompensasi termal**: Pengukuran suhu digunakan untuk mengoreksi keluaran sensor yang dipengaruhi oleh suhu lingkungan.  
- **Sintesis gambar**: Data dari **radar, optik, dan inframerah** dikombinasikan untuk menghasilkan citra yang lebih komprehensif.

<img src="fusion.png" width="450"/>

## 9. Estimasi dalam Pengukuran
Dengan kemajuan komputasi, akurasi pengukuran dapat ditingkatkan melalui **teknik estimasi**.  
Metode yang digunakan:  
1. **Perataan sederhana atau penyaringan frekuensi rendah** untuk menghilangkan fluktuasi acak.  
2. **Penyaringan Wiener**, teknik lanjutan untuk meningkatkan akurasi pengukuran tanpa meningkatkan kualitas sensor secara fisik.
